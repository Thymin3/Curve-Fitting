# ELISA Concentration computation with 4 Parameter Logistic curve

### Background Info


source: https://www.myassays.com/four-parameter-logistic-regression.html

### Importing relevant libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import warnings
from sklearn.metrics import r2_score
import pandas as pd
import openpyxl
```

### Curve fitting + computation of standard curve

```python
warnings.filterwarnings("ignore")

#### computation of standard concentrations ####  
st_start = 20000            # highest standard concentration
number_of_st_values = 12    # number of different standard concentrations
dilution_factor = 2         # dilution factor between standards
st_values = []              # list of standard concentrations
for i in range(number_of_st_values):
    if len(st_values) == 0:
        st_values.append(st_start)
    else:
        st_values.append(st_start/dilution_factor)
        st_start = st_start/dilution_factor   
#################################################

x = np.array(st_values)     # using standard concentrations as x-values

y = np.array([3.0768, 2.9526, 2.068, 1.0887, 0.5681, 0.2883, 0.1667, 0.0989, 0.0706, 0.0631, 0.0466, 0.0367]) #example data for measured ODs of standard curve for y-values


def FourPLFunction(x, a, b, c, d):
    '''Determines the "y" variable by inserting the already determined x, a, b, c, d parameters of a four parameter logistic curve.
    Background info to 4PL regression: https://www.myassays.com/four-parameter-logistic-regression.html'''
    return d+(a-d)/(1+ (x/c)**b) 

#### computation of 4PL parameters a, b ,c, d with the help of x and y standard values ####
parameters_optimal, parameters_cov = optimize.curve_fit(FourPLFunction, x, y, p0=[y[-1], 1, 1, y[0]]) #cov = covariance; p0 = initial values for curve fit

a = parameters_optimal[0] # min
b = parameters_optimal[1] # inflection point
c = parameters_optimal[2] # hills slope of the curve (=steepness at point b)
d = parameters_optimal[3] # max
###########################################################################################

x_model = np.linspace(min(x), max(x)+100000, 1000) # used to model standard curve
y_model = FourPLFunction(x_model, a, b, c, d) # used to model standard curve

y_predicted = FourPLFunction(x, a, b, c, d) # used to compute R squared value
```

### Graphics section

```python
print("a = ", a, " min value", "\nb = ", b, " inflection point", "\nc = ", c, " steepness at point b", "\nd = ",d, " max value" "\n") 

print("R^2 = ", r2_score(y, y_predicted)) # computation of R^2

plt.scatter(x, y, color = "red", marker = "o", label='Raw data')
plt.title("ELISA IFNy readout")
plt.plot(x_model, y_model, label = "Standard curve")
plt.xlabel("IFNy concentration [pg/ml]")
plt.ylabel("OD")
plt.legend(loc = "lower right")
plt.xscale('log')
plt.xlim([0, 100000])
plt.show()
```
